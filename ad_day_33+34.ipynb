```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic = pd.read_csv(r"/content/Titanic.csv")
```

```python
df=pd.get_dummies(titanic,dtype='int')
df.head()
```

```python
fequent_itemsets=apriori(df,min_support=0.1,use_colnames=True)
fequent_itemsets
```

```python
rules=association_rules(fequent_itemsets,metric='lift',min_threshold=0.7)
rules
```
